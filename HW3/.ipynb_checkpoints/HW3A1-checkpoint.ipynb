```python
import numpy as np
import pylab as pl
%pylab inline 
```

```python
myDist = {}
distributions = ['pois', 'norm', 'chisq', 'gamma', 'logistic', 'binom']
myMean = 100
```

```python
# creating sample distributions with 100 key value pairs per distribution

for d in distributions:
    myDist[d] = {}

mySize = 20000/(np.array(range(1,101))+10)

for n in mySize:
    myDist['pois'][n] = np.random.poisson(myMean, size=n)
    myDist['norm'][n] = np.random.normal(myMean, size=n)
    myDist['chisq'][n] = np.random.chisquare(myMean, size=n)
    myDist['gamma'][n] = np.random.gamma(myMean, size=n)
    myDist['logistic'][n] = np.random.logistic(myMean, 1, size=n)
    myDist['binom'][n] = np.random.binomial(2*myMean, .5, size=n)
    
```

```python
#plotting all 6 sample distributions

fig_mu_N = pl.figure(figsize=(10,10))
axpois_mu_n = fig_mu_N.add_subplot(1,1,1)
axnorm_mu_n = fig_mu_N.add_subplot(1,1,1)
axchisq_mu_n = fig_mu_N.add_subplot(1,1,1)
axgamma_mu_n = fig_mu_N.add_subplot(1,1,1)
axlogistic_mu_n = fig_mu_N.add_subplot(1,1,1)
axbinom_mu_n = fig_mu_N.add_subplot(1,1,1)

for nn in mySize:
    axpois_mu_n.plot(nn, myDist['pois'][nn].mean(),'o',color='b',label='Chi-Square')
    axnorm_mu_n.plot(nn, myDist['norm'][nn].mean(),'o',color='k')
    axchisq_mu_n.plot(nn, myDist['chisq'][nn].mean(),'o',color='g')
    axgamma_mu_n.plot(nn, myDist['gamma'][nn].mean(),'o',color='m')
    axlogistic_mu_n.plot(nn, myDist['logistic'][nn].mean(), 'o', color='r')
    axbinom_mu_n.plot(nn, myDist['binom'][nn].mean(), 'o', color='y')
    axchisq_mu_n.plot([min(mySize), max(mySize)],[myMean,myMean], 'k') 
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')

pl.legend(distributions)
pl.show()
```

#Q: Describe the behavior you see in the plots - do they look as you expected? why?
#A: as the sample size increase, the sample mean gets closer and closer to the population mean. (Central Limit Theorem)

```python
#histogram of the mean of all samples 

means = {}

for d in distributions:
    means[d] = {}

for d in means:
    for s in myDist[d]:
        means[d][s] = myDist[d][s].mean()
        
allMeans=[]

for d in means:
    for i in means[d]:
        allMeans.append(means[d][i])
pl.figure(figsize=(10,10))
pl.hist(allMeans,bins=30)
pl.xlabel('sample mean')
pl.ylabel('N')
```
